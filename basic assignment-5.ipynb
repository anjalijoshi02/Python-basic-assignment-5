#### Question 8 : How do you print "pretty print" dictionary values using which module and function?

We can pretty print a dictionary using three functions:    
1. by using pprint function of pprint module.    
2. by using dumps() method of json module.    
3. by using dumps() methods of yaml module.

```python
import pprint
import json 
import yaml
d=[{'name':'abc','age':24,'residence':{'country':'India','city':'New Delhi'}},
   {'name':'pqr','age':25,'residence':{'Country':'India','city':'Kolkata'}},
   {'name':'xyz','age':26,'residence':{'country':'India','city':'Kanpur'}}]

print("printing using print function \n",d)
print('-'*100)
print("printing using pprint function \n")
pprint.pprint(d)
print('-'*100)
dump=json.dumps(d,indent=3)
print("Printing using dumps of json module \n",dump)
print('-'*100)
dump1=yaml.dump(d)
print("printing using dumps of yaml module \n",dump1)
```
